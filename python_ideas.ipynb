### Variable Types

#### Strings

```python
# Strings can be stuck together.
my_string = 'Py' 'thon'
print('my_string:', my_string)
# Strings are immutable, but can be joined to variables...
print(my_string + ' is fun.')
# ... and new strings can be created through slicing.
print('J' + my_string[1:])
# Strings have special characters which can be dealt with.
print('Bad string location:', 'C:\new\location')
print('Good (raw) string location:', r'C:\new\location')
# Fast strings can be used to report variable outputs.
a, b = 0, 1  # Note the multiple assignment!
print(f'Fast strings make variables easy. a: {a}, b: {b}')
```

#### Lists

```python
# Lists can be created through list comprehensions.
cubes = [i ** 3 for i in range(1, 6, 1)]
print('cubes:', cubes)
# Lists are mutable.
cubes_mutated = cubes.copy()
cubes_mutated[1] = 'eight'
print('cubes_mutated:', cubes_mutated)
# Lists can be added to.
print('Cubes added to:', cubes + [216, 343])
# Lists also have the append() method for sinlge additions.
cubes.append(6 ** 3)
print('Cubes appended to:', cubes)
# Lists may contain lists.
squares = [i ** 2 for i in range(1, 6)]
seqs = [squares, cubes]
print('seqs:', seqs)
print('First list, second element:', seqs[0][1])
# Elements can be inserted at a specified position.
squares.insert(1, 42)
print('Squares insert:', squares)
# Elements can also be removed.
squares.remove(42)
print('Squares with 42 removed:', squares)
# Elements can be returned and removed using pop().
play_list = ['John', ' Michael', 'Guido']
play_list.pop(1)
print('Popped list:', play_list)
# A list can be cleared. Use del the_list to remove it.
# Also, del the_list[:] also clears the list.
play_list.clear()
print('Cleared list:', play_list)
# Elements in a list can be counted.
play_list = [0, 0, 0, 1, 2, 2]
print('Count zeros:', play_list.count(0))
# Lists can be sorted.
play_list.sort(reverse=True)
print('Reversed list:', play_list)
# Reversing (without sorting) is also possible.
play_list.reverse()
print('Alternatively reversed list:', play_list)
# List can be copied. new = old[:] also works.
new_play_list = play_list.copy()
new_play_list[0] = 'New'
print('New list:', new_play_list)
print('Old list:', play_list)
```

##### Lists as Stacks

Stacks are a last-in, first-out method of storing and using data.

```python
names = ['John', 'Michael']
names.append('Guido')
names.pop()
```

##### Lists as Queues

These are a first-in, first-out method of storing and using data. The standard library has *deque()* in **collections** which allows queues to be quick.

```python
from collections import deque
q = deque(['Eric', 'John', 'Michael'])
q.append('Terry')
q.popleft()
```

#### Tuples and Sequences

Tuples are immutable data structures which can be nested with different types of object. Some useful techniques are outlined below.

```python
empty = ()
singleton = 'hello',  # Note the trailing comma.
print('Lengths of empty and singleton are '
      f'{len(empty)} and {len(singleton)}'
      ', respectively.')
t = 12345, 54321, 'hello!'  # Tuple packing.
x, y, z = t  # Tuple unpacking.
print(f'The tuple t: {t}')
u = t, (1, 2, 3, 4, 5)
print(f'The tuple u: {u}')
print(f'First element of u: {u[0]}')
v = u, ([0, 1, 2], [2, 1, 0])
print(f'The tuple v: {v}')
```

##### Sets

Sets are unordered collections of unique elements. The canbe assigned with *set()* or braces (*i.e.* {}).

```python
basket = {'apple', 'orange', 'apple', 'pear', 'orange', 'banana'}
print(f'Basket: {basket}')
print(f'Orange check: {"orange" in basket}')
a = set('abracadabra')
b = set('alazcazam')
print(f'Unique letters in a: {a}')
print(f'Letters in a but not b: {a - b}')
print(f'Letters in a or b, or both: {a | b}')
print(f'Letters in a or b, but not both: {a ^ b}')
print(f'Letters in a and b: {a & b}')
# Set comprehensions also exist.
s = {x for x in 'abracadabra' if x not in 'abc'}
print(f's: {s}')
```

##### Dictionaries

Dictionaries are collections of unique key-value pairs. Keys can be made from any immutable type (*e.g.* tupes are OK, lists are not).

```python
empty_dict0 = {}
empty_dict1 = dict()  # Alternative.
tel = {'jack': 4098, 'sape': 4139}
tel0 = dict([('jack', 4098), ('sape', 4139)])  # Alternative.
tel1 = dict(jack=4098, sape=4139)  # For simple strings.
tel['guido'] = 4127
print(f'The dictionary called tel: {tel}')
del tel['jack']
tel['irv'] = 4127
print(f'Edited version of tel: {tel}')
print(f'Keys from tel: {list(tel)}')
print(f'Sorted keys from tel: {sorted(tel)}')
print(f'Check if a key is present: {"guido" in tel}')
# Dictionary comprehension exist.
dc = {x: x**2 for x in (2, 4, 6)}
print(f'Comprehension dictionary: {dc}')
```

### Control Flows

#### while

When a loop should repeat until some criterion is satisfied.

```python
a, b = 0, 1
while a < 1e3:
    print(a, end=',')
    a, b = b, a + b
```

#### if

To check against some reference value.

```python
x = float(input('Please supply a real number: '))
if x < 0:
    print('The number is negative.')
elif x % 10 == 0:  # Optional.
    print('The number is positive and wholly divisible by ten.')
else:  # Optional.
    print('The number is positivive and not wholly divisible by ten.')
```

#### for

To loop through all the values.

```python
colours = ['red', 'blue', 'green']
for i, colour in enumerate(colours):
    print(i, '--->', colour)
```

For loops can use lists to edit dictionaries. It is best to iterate over a copy of a list to avoid complications from overwriting while checking.

```python
users = {'Guido': 'active',
         'Raymond': 'inactive',
         'Laura': 'active'}
for user, status in users.copy().items():
    if status == 'inactive':
        del users[user]
print(users)
```

Dictionaries can be looped through in an elegant manner using the following appoach. Note that the dictionary **tel** was defined earlier.

```python
for name, number in tel.items():
    print(name, number)
```

#### range

Allows a sequence of numbers to be created.

```python
list(range(0, 6, 2))  # Start, stop, by.
```

#### break

This can be used to break out of the innermost loop of a for or while loop.

```python
for n in range(2, 10):
    for x in range(2, n):
        if n % x == 0:
            print(f'{n} equals {x} * {n // x}')
            break
        else:
            print(f'{n} is a prime number.')
```

##### zip

A zip can be used to join two or more sequences.

```python
questions = ['name', 'quest', 'favourite colour']
answers = ['lancelot', 'the holy grail', 'blue']
for q, a in zip(questions, answers):
    print(f'What is your {q}? It is {a}.')
```

##### Assignmnet Inside Experessions (The Walrus! :=)

Assignments can take place inside expressions in Python 3.8+.

```python
my_list = [0, 1, 2, 3, 4, 5]
if (n := len(my_list)) > 3:
    print(f'The list is too long. It has {n} elements.')
```

### continue

A continue forces a loop onto the next iteration, ignoring any subsequent code after the **continue**.

```python
for n in range(2, 10):
    if n % 2 == 0:
        print(f'{n} is even.\n')
        continue
    else:
        print(f'{n} is odd.')
    print('Boo! Still running! ;-P\n')
```

#### Functions

```python
def fib(n):
    """Return a Fibonacci series up to n."""
    result = []
    a, b = 0, 1
    while a < n:
        result.append(a)
        a, b = b, a + b
    return result

fib_1e3 = fib(1000)
fib_1e3
```

#### pass

This can be used as a placeholder in loops and functions to allow code to run without any further code.

```python
def f():
    """A function with only a pass."""
    pass

f()
```

##### Default Arguments

```python
def ask_ok(prompt, retries=4, reminder='Try again.'):
    while True:
        ok = input(prompt)
        if ok in ('y', 'ye', 'yes'):
            return True
        if ok in ('n', 'no', 'nop', 'nope'):
            return False
        retries = retries - 1
        if retries < 0:
            raise ValueError('invalid user response')
        print(reminder)


ask_ok('Do you want to quit', 'yes', 'yes or no')
```

##### Keyword Arguments

```python
# *name -> tuple
# **name -> dictionary
def cheeseshop(kind, *args, **kws):
    print(f'Do you have any {kind}?')
    print(f'I\'m sorry, we\'re all out of of {kind}.')
    for arg in args:
        print(arg)
    for kw in kws:
        print(f'{kw}: {kws[kw]}')


cheeseshop('Limburger', 'It\'s very runny, sir.',
           'It\'s really very, VERY runny, sir.',
           shopkeeper='Michael Palin.',
           client='John Cleese.',
           sketch='Cheese Shop Sketch.')
```

##### Special Parameters

Positional only can be enforced using forward slash and asterisk.

```python
def f(pos1, pos2, /, pos_or_kw, *, kw1, kw2):
    pass


f(0, 1, 2, kw1=3, kw2=4)  # 2 could be pos_or_kw=2
```

This can make it possible to have inputs that would otherwise conflict with arguments.

```python
def foo(name, /, **kwds):
    return 'name' in kwds


foo(1, **{'name': 2})
```

##### Arbitrary Arguments Lists

Tuple input arguments can be used to *e.g.* join strings together.

```python
def concat(*args, sep=', '):
    return sep.join(args)


concat('earth', 'mars', 'venus')
```

Dictionaries can also be unpacked using tuple input arguments to a function.

```python
def parrot(voltage, state='a stiff', action='voom'):
    print(f'This parrot wouldn\'t {action}', end=' ')
    print(f'if you put {voltage} volts through it.', end=' ')
    print(f'\'E\'s {state}.')


d = {'voltage': 'four million', 'state': 'bleedin\' demised',
     'action': 'voom'}
parrot(**d)
```
