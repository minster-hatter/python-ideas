### Variable Types

#### Strings

```python
# Strings can be stuck together.
my_string = 'Py' 'thon'
print('my_string:', my_string)
# Strings are immutable, but can be joined to variables...
print(my_string + ' is fun.')
# ... and new strings can be created through slicing.
print('J' + my_string[1:])
# Strings have special characters which can be dealt with.
print('Bad string location:', 'C:\new\location')
print('Good (raw) string location:', r'C:\new\location')
# Formatted strings can be used to report variable outputs.
a, b = 0, 1  # Note the multiple assignment!
print(f'Formatted strings make variables easy. a: {a}, b: {b}')
```

#### Lists

```python
# Lists can be created through list comprehensions.
cubes = [i ** 3 for i in range(1, 6, 1)]
print('cubes:', cubes)
# Lists are mutable.
cubes_mutated = cubes.copy()
cubes_mutated[1] = 'eight'
print('cubes_mutated:', cubes_mutated)
# Lists can be added to.
print('Cubes added to:', cubes + [216, 343])
# Lists also have the append() method for sinlge additions.
cubes.append(6 ** 3)
print('Cubes appended to:', cubes)
# Lists may contain lists.
squares = [i ** 2 for i in range(1, 6)]
seqs = [squares, cubes]
print('seqs:', seqs)
print('First list, second element:', seqs[0][1])
# Elements can be inserted at a specified position.
squares.insert(1, 42)
print('Squares insert:', squares)
# Elements can also be removed.
squares.remove(42)
print('Squares with 42 removed:', squares)
# Elements can be returned and removed using pop().
play_list = ['John', ' Michael', 'Guido']
play_list.pop(1)
print('Popped list:', play_list)
# A list can be cleared. Use del the_list to remove it.
# Also, del the_list[:] also clears the list.
play_list.clear()
print('Cleared list:', play_list)
# Elements in a list can be counted.
play_list = [0, 0, 0, 1, 2, 2]
print('Count zeros:', play_list.count(0))
# Lists can be sorted.
play_list.sort(reverse=True)
print('Reversed list:', play_list)
# Reversing (without sorting) is also possible.
play_list.reverse()
print('Alternatively reversed list:', play_list)
# List can be copied. new = old[:] also works.
new_play_list = play_list.copy()
new_play_list[0] = 'New'
print('New list:', new_play_list)
print('Old list:', play_list)
```

#### Tuples and Sequences

Tuples are immutable data structures which can be nested with different types of object. Some useful techniques are outlined below.

```python
empty = ()
singleton = 'hello',  # Note the trailing comma.
print('Lengths of empty and singleton are '
      f'{len(empty)} and {len(singleton)}'
      ', respectively.')
t = 12345, 54321, 'hello!'  # Tuple packing.
x, y, z = t  # Tuple unpacking.
print(f'The tuple t: {t}')
u = t, (1, 2, 3, 4, 5)
print(f'The tuple u: {u}')
print(f'First element of u: {u[0]}')
v = u, ([0, 1, 2], [2, 1, 0])
print(f'The tuple v: {v}')
```

##### Sets

Sets are unordered collections of unique elements. The canbe assigned with *set()* or braces (*i.e.* {}).

```python
basket = {'apple', 'orange', 'apple', 'pear', 'orange', 'banana'}
print(f'Basket: {basket}')
print(f'Orange check: {"orange" in basket}')
a = set('abracadabra')
b = set('alazcazam')
print(f'Unique letters in a: {a}')
print(f'Letters in a but not b: {a - b}')
print(f'Letters in a or b, or both: {a | b}')
print(f'Letters in a or b, but not both: {a ^ b}')
print(f'Letters in a and b: {a & b}')
# Set comprehensions also exist.
s = {x for x in 'abracadabra' if x not in 'abc'}
print(f's: {s}')
```

##### Dictionaries

Dictionaries are collections of unique key-value pairs. Keys can be made from any immutable type (*e.g.* tupes are OK, lists are not).

```python
empty_dict0 = {}
empty_dict1 = dict()  # Alternative.
tel = {'jack': 4098, 'sape': 4139}
tel0 = dict([('jack', 4098), ('sape', 4139)])  # Alternative.
tel1 = dict(jack=4098, sape=4139)  # For simple strings.
tel['guido'] = 4127
print(f'The dictionary called tel: {tel}')
del tel['jack']
tel['irv'] = 4127
print(f'Edited version of tel: {tel}')
print(f'Keys from tel: {list(tel)}')
print(f'Sorted keys from tel: {sorted(tel)}')
print(f'Check if a key is present: {"guido" in tel}')
# Dictionary comprehension exist.
dc = {x: x**2 for x in (2, 4, 6)}
print(f'Comprehension dictionary: {dc}')
```

#### Functions

```python
def fib(n):
    """Return a Fibonacci series up to n."""
    result = []
    a, b = 0, 1
    while a < n:
        result.append(a)
        a, b = b, a + b
    return result

fib_1e3 = fib(1000)
fib_1e3
```

##### Default Arguments

```python
def ask_ok(prompt, retries=4, reminder='Try again.'):
    while True:
        ok = input(prompt)
        if ok in ('y', 'ye', 'yes'):
            return True
        if ok in ('n', 'no', 'nop', 'nope'):
            return False
        retries = retries - 1
        if retries < 0:
            raise ValueError('invalid user response')
        print(reminder)


ask_ok('Do you want to quit', 'yes', 'yes or no')
```

##### Keyword Arguments

```python
# *name -> tuple
# **name -> dictionary
def cheeseshop(kind, *args, **kws):
    print(f'Do you have any {kind}?')
    print(f'I\'m sorry, we\'re all out of of {kind}.')
    for arg in args:
        print(arg)
    for kw in kws:
        print(f'{kw}: {kws[kw]}')


cheeseshop('Limburger', 'It\'s very runny, sir.',
           'It\'s really very, VERY runny, sir.',
           shopkeeper='Michael Palin.',
           client='John Cleese.',
           sketch='Cheese Shop Sketch.')
```

```python
import math
# Control the number of decimal places, n, with foo:.nf
print(f'The value of pi as approximately {math.pi:.3f}')
# Control minimum width, w, of a field with {foo:.wd}.
table = dict(Sjoerd=4127, Jack=4098, Dcab=7678)
for name, number in table.items():
    print(f'{name:10} --> {number:10d}')
    # repr() can be used to represent strings in output stings.
str_repr = 'output'
print(f'Here is an output with str: {str_repr!s}')
print(f'Here is an output with repr: {str_repr!r}')
# Manual string formatting with justification (rjust(), ljust(), center()).
for x in range(1, 11):
    print(repr(x).rjust(2), repr(x * x).rjust(3), end=' ')
    print(repr(x * x * x).rjust(4))
# str.zfill() can be used to pad a string, e.g.
print(f'Padded pi: {"3.14".zfill(7)}')
```

#### Reading and Writing Files

A *with* statement is best used to read files as it automatically closes the connection to the file on completing the statement.

```python
import os

# Append mode, a, either appends to or creates file if none exists.
file_string = ('This is the first line.\n'
               'This is the second line.\n'
               'Welcome to the text document!\n')
with open('text_doc.txt', 'a') as f:
    f.write(file_string)

with open('text_doc.txt') as f:
    read_file = f.read()
print(read_file)

with open('text_doc.txt') as f:
    print(f'Readline: {f.readline()}')

with open('text_doc.txt') as f:
    lines = f.readlines()
    words = []
    for n, line in enumerate(lines):
        print(f'Line {n}: {line}', end='')
        words.append(line.split())
    print(f'Words: {words}')

os.remove('text_doc.txt')
```

##### Structured Data and json

Structured data can be written and read using json format.

```python
import json
import os

x = [1, 'simple', 'list']
print(f'x in json format: {json.dumps(x)}')
# Serialize (write) x to a text file.
with open('json_file.txt', 'a') as f:
    json.dump(x, f)

# Deserialize (load) x back from the file.
with open('json_file.txt') as f:
    data = json.load(f)
print(f'Recovered data: {data}')

os.remove('json_file.txt')
```

### Scopes and Namespaces

Namespaces map names to objects. Two modules can have the same name for different functions as they have different namespaces. 

Note that names can be removed from namespaces with a *del module.attrbute* statement. 

The keywords *global* and *nonlocal* exist to assign a name to the approriate scope. For example:

```python
def scope_test():
    def do_local():
        spam = 'local spam'

    def do_nonlocal():
        spam = 'nonlocal spam'

    def do_global():
        global spam
        spam = 'global spam'

    spam = 'test spam'
    do_local()
    print(f'After local assignment: {spam}')
    do_nonlocal()
    print(f'After nonlocal assignment: {spam}')
    do_global()
    print(f'After global assignment: {spam}')


scope_test()
print(f'In global scope: {spam}')
```

#### Classes and Instance Variables

Variables can be shared or unique between class instances.

```python
class Dog:
    
    kind = 'canine'  # Class variable shared by all instances.
    
    def __init__(self, name):
        self.name = name  # Instance variable unique to an instnace.


d = Dog('Fido')
e = Dog('Buddy')
print(f'd.kind and e.kind: {d.kind, e.kind}')
print(f'd.name and e.name: {d.name, e.name}')
```

This has implications with mutable objects when the class variable is used poorly. For example:

```python
class Dog:
    
    tricks = []
    
    def __init__(self, name):
        self.name = name
    
    def add_trick(self, trick):
        self.tricks.append(trick)


d = Dog('Fido')
e = Dog('Buddy')
d.add_trick('roll over')
e.add_trick('play dead')
print(f'd.tricks: {d.tricks}')
```

This can be avoided through the following coding:

```python
class Dog:

    def __init__(self, name):
        self.name = name
        self.tricks = []

    def add_trick(self, trick):
        self.tricks.append(trick)


d = Dog('Fido')
e = Dog('Buddy')
d.add_trick('roll over')
e.add_trick('play dead')
print(f'd.tricks: {d.tricks}\ne.tricks: {e.tricks}')
```

### Errors and Exceptions

Python has several built-in exceptions. These include, but are not limited to, ZeroDvisionError, NameError, and TypeError. Specific exceptions can be handled with a *try* statement. Errors that do not match the stated error(s) will be a handled exception (as there is no specified 'handler').

```python
try:
    x = int('A string!')
except ValueError:
    print('That was not a valid number. Try again.')
# Note that multiple exceptions can be specifed as follows:
# except (TypeError, NameError ...)
```

A *raise* statement raises an exception.

```python
raise NameError('This is the raised name error.')
```

Exceptions work for a class if it is the same class, or a base class thereof, but not *vice versa*.

```python
class B(Exception):
    pass


class C(B):
    pass


class D(C):
    pass


for cls in [B, C, D]:
    try:
        raise cls()
    except D:
        print('D')
    except C:
        print('C')
    except B:
        print('B')  

# The result would be B, B, B if the last section was first.
```

#### User-Defined Exceptions

New exceptions can be created as classes. These should almost always be derived from the *Exception* class. For example ...

```python
class Error(Exception):
    """Base class for exceptions in this module."""
    pass


class InputError(Error):
    """Exception raised for input errors.
    
    Attributes:
        expression -- input expression in which the error occured.
        message -- explanation of the error
    """
    def __inti__(self, expression, message):
        self.expression = expression
        self.message = message


class TransitionError(Error):
    """Raised when an operation attempts a state transition that
    is not allowed.
    
    Attributes:
        previous -- state at beginning of transition
        next -- attempt new state
        message -- explanation of why the specific transition is not allowed
    """
    def __init__(self, previous, next, message):
        self.previous = previous
        self.next = next
        self.message = message
```

#### Defining Clean-up Actions

To deal with exceptions occuring, *try* can be used in conjunction with *finally*.

```python
def divide(x, y):
    try:
        result = x / y
    except ZeroDivisionError:
        print('Division by zero!')
    else:
        print(f'The result is {result}.')
    finally:
        print('Executing the finally clause.\n')


divide(2, 1)
divide(2, 0)
divide('2', '1')
```
